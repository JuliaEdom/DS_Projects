# Успешность игр

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation).   

Нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.  

Доступны данные до 2016 года, в которых попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**Описание данных**

        Name — название игры
        Platform — платформа
        Year_of_Release — год выпуска
        Genre — жанр игры
        NA_sales — продажи в Северной Америке (миллионы проданных копий)
        EU_sales — продажи в Европе (миллионы проданных копий)
        JP_sales — продажи в Японии (миллионы проданных копий)
        Other_sales — продажи в других странах (миллионы проданных копий)
        Critic_Score — оценка критиков (максимум 100)
        User_Score — оценка пользователей (максимум 10)
        Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

**План проекта**

```python
# поле year_of_release содержит 1,6% пропусков. Количество пропущенных значений некритично, а год релиза значим для анализа  
# удалим эти строки и сменим формат на int - для возможности работы с числами данного столбца

data = data[~data.year_of_release.isna()]
data.year_of_release = pd.to_numeric(data.year_of_release, downcast='integer') 
```

```python
# заменим значения tbd на nan и преобразуем формат столбца user_score в числовой (числовой формат необходим 
# для дальнейшего анализа)
# заменим nan в рейтинге на 'нет рейтинга' 

data.loc[data.user_score == 'tbd', 'user_score'] = np.nan
data.user_score = pd.to_numeric(data.user_score)
data.rating = data.rating.fillna('нет рейтинга')
```

```python
# найдем годы запуска новых платформ, а также годы прекращения релизов на них 
dict = {}
h = 0

for platf in data_pl.platform.unique():
    dict[platf] = [data_pl[data_pl.platform == platf].year_of_release.min(), \
                   data_pl[data_pl.platform == platf].year_of_release.max(),   \
                   data_pl[data_pl.platform == platf].year_of_release.max() -    \
                        data_pl[data_pl.platform == platf].year_of_release.min()]
    
    h += data_pl[data_pl.platform == platf].year_of_release.max() -    \
                        data_pl[data_pl.platform == platf].year_of_release.min()

pl_stat = pd.DataFrame(dict)
pl_stat.index = ['start', 'end', 'life_time']
year_end_unique = sorted(list(pl_stat.T.end.unique()))[:-1]
year_start_unique = sorted(list(pl_stat.T.start.unique()))

h/31
```

В среднем платформа живет порядка 8 лет

```python
# посмотрим, сколько в среднем проходило между годами запуска новых платформ

s=[]

for i in range(1, len(year_start_unique)):
    s.append(year_start_unique[i] - year_start_unique[i-1])
              
sum(s) / len(s)
```

```python
# посмотрим, сколько в срднем проходило между годами прекращения работы старых платформ

m=[]

for i in range(1, len(year_end_unique)):
    m.append(year_end_unique[i] - year_end_unique[i-1])
              
sum(m) / len(m)
```

В среднем каждые 2 года появлялись новые платформы и прекращали существование старые

## 5. Проверка гипотез
- средние пользовательские рейтинги платформ Xbox One и PC одинаковые
- средние пользовательские рейтинги жанров Action и Sports разные

Нулевую гипотезу в обоих случаях сформулируем единственным возможным образом - равенство средних, альтернативную - неравенство средних. Уровень статистической значимости зададим 0,05

```python
xone = data.query('platform == "XOne" and year_of_release <= 2015 and year_of_release >= 2013').user_score
pc = data.query('platform == "PC" and year_of_release <= 2015 and year_of_release >= 2013').user_score
```

```python
display(xone.describe().to_frame().T)
display(pc.describe().to_frame().T)
```

```python
# оценим близость распределения выборок к нормальному распределению

plt.figure(figsize=(16,4))
ax1 = plt.subplot(121)
stats.probplot(xone.dropna(), dist="norm", plot=plt)
ax1.set_title('Q-Q Prob Plot. Распределение оценок XOne')
ax2 = plt.subplot(122)
stats.probplot(pc.dropna(), dist="norm", plot=plt)
ax2.set_title('Q-Q Prob Plot. Распределение оценок PC')
plt.show()
```

Распределение оценок по обеим платформам близко к нормальному со смещением влево

```python
res = stats.ttest_ind(xone.dropna(), pc.dropna(), equal_var = False)
print('p-значение: ', res.pvalue)
```

Значение pvalue выше выбранного уровня статистической значимости. Не отвергаем нулевую гипотезу - средние пользовательские рейтинги платформ Xbox One и PC одинаковые 

```python
action = data.loc[(data['genre'] == 'Action') & (~data['user_score'].isna() & (data.year_of_release >= 2013) & \
                                                     (data.year_of_release <= 2015)), 'user_score']
sports = data.loc[(data['genre'] == 'Sports') & (~data['user_score'].isna() & (data.year_of_release >= 2013) & \
                                                     (data.year_of_release <= 2015)), 'user_score']
```

```python
display(action.describe().to_frame().T)
display(sports.describe().to_frame().T)
```

```python
# оценим близость распределения выборок к нормальному распределению

plt.figure(figsize=(16,4))
ax1 = plt.subplot(121)
stats.probplot(action, dist="norm", plot=plt)
ax1.set_title('Q-Q Prob Plot. Распределение оценок по жанру Action')
ax2 = plt.subplot(122)
stats.probplot(sports, dist="norm", plot=plt)
ax2.set_title('Q-Q Prob Plot. Распределение оценок по жанру Sports')
plt.show()
```

Распределение оценок по обоим жанрам также близко к нормальному со смещением влево

```python
res_ = stats.ttest_ind(action, sports, equal_var = False)
print('p-значение: ', res_.pvalue)
```

Отвегаем нулевую гипотезу. Средние пользовательские рейтинги жанров Action и Sports неодинаковые

Таким образом, мы приняли обе исходные гипотезы 
  - средние пользовательские рейтинги платформ Xbox One и PC одинаковые
  - средние пользовательские рейтинги жанров Action и Sports разные
