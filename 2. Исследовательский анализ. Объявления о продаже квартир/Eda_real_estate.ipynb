# Исследование объявлений о продаже квартир

Необходимо выявить влияние отдельных признаков объектов недвижимости на их рыночную стоимость. 

Исходные данные - архив объявлений сервиса Яндекс.Недвижимость о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет.  

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

**Описание данных**

    airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
    balcony — число балконов
    ceiling_height — высота потолков (м)
    cityCenters_nearest — расстояние до центра города (м)
    days_exposition — сколько дней было размещено объявление (от публикации до снятия)
    first_day_exposition — дата публикации
    floor — этаж
    floors_total — всего этажей в доме
    is_apartment — апартаменты (булев тип)
    kitchen_area — площадь кухни в квадратных метрах (м²)
    last_price — цена на момент снятия с публикации
    living_area — жилая площадь в квадратных метрах (м²)
    locality_name — название населённого пункта
    open_plan — свободная планировка (булев тип)
    parks_around3000 — число парков в радиусе 3 км
    parks_nearest — расстояние до ближайшего парка (м)
    ponds_around3000 — число водоёмов в радиусе 3 км
    ponds_nearest — расстояние до ближайшего водоёма (м)
    rooms — число комнат
    studio — квартира-студия (булев тип)
    total_area — площадь квартиры в квадратных метрах (м²)
    total_images — число фотографий квартиры в объявлении

## 2. Предобработка данных

```python
# дата публицации содержит в формате время, которое по фактту отсутствует, избавимся от неинформативных 0
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format = '%Y-%m-%d')
data
```

```python
# создадим новую колонку с локацией, где будет только имя собственное
# таким образом исключим неявные дубликаты, также условно сольем деревни и поселки с одинаковыми названиями - 
# это один и тот же населенный пункт или соседние населенные пункты

def loc_cor(row):
    if row['locality_name'] == 'Санкт-Петербург' or pd.isna(row['locality_name']):
        return np.nan
    elif row['locality_name']:
        t = []
        s = str(row['locality_name'])
        for x in s.split():
             if x[0] == x[0].upper():
                t.append(x)
        m = ' '.join(t)
        return m
            
data['locality_cor'] = data.apply(loc_cor, axis=1)
```

```python
di = {}
for loc in data.locality_cor.unique():
    if pd.isna(loc):
        di[loc] = np.nan
    elif data[(data.locality_cor == loc)&(loc != 'Санкт-Петербург')].cityCenters_nearest.median():
        di[loc] = data[data.locality_cor == loc].cityCenters_nearest.median()
```

```python
# если в населенном пункте есть объекты, для которых отсутствует расстояние до центра, но есть объекты с этим расстоянием -
# заполним пропуски медианой
for x in di:
    data.loc[(data.locality_cor == x)&(data.cityCenters_nearest.isna()), 'cityCenters_nearest'] = di[x]
```

```python
geolocator = Nominatim(user_agent="xxx@gmail.com")
pit_cent = geolocator.geocode("Санкт-Петербург невский проспект, 85")

dict_coordinates = {}
# локации, для которых отсутствует расстояние до центра
data[data.cityCenters_nearest.isna()].locality_cor.unique()
#создадим словарь - локация-расстояние до центра Санкт-Петербурга (условно примем адрес Московсого вокзала как
#центральную точку
for x in data[data.cityCenters_nearest.isna()&(data.locality_cor != 'Санкт-Петербург')].locality_cor.unique():
    loc = 'Ленинградская область ' + str(x)
    if geolocator.geocode(loc):
        location = geolocator.geocode(loc)
        dict_coordinates[x] = geodesic((pit_cent.latitude, pit_cent.longitude), 
                                       (location.latitude, location.longitude)).meters
```

```python
for x in dict_coordinates:
    data.loc[(data.locality_cor == x)&(data.cityCenters_nearest.isna()), 'cityCenters_nearest'] = \
                                                                                         dict_coordinates[x]
def f(row):
    if row['locality_name'] == 'Санкт-Петербург':
        return 'Санкт-Петербург'
    
data.loc[data.locality_cor.isna(), 'locality_cor'] = data[data.locality_cor.isna()].apply(f, axis=1)
```

```python
#удалим строки, для которых одновременно нет информации по локации и по расстоянию от центра - расположение данных
# объектов неизвестно
ind_to_del = data[(data.locality_name.isna())&(data.cityCenters_nearest.isna())].index

data.drop(ind_to_del, inplace=True)
```

```python
data['km_to_center'] = round(data['cityCenters_nearest']/1000)
```

```python
# высота потолков более 20 м - явно смещение точки, ниже разделим значение на 10 
data[data.ceiling_height > 20]
```

```python
data.loc[data.ceiling_height > 20, 'ceiling_height'] = data.loc[data.ceiling_height > 20, 'ceiling_height'] / 10
```

```python
# заполним пропуски по высоте потолков медианным значением для центра Санкт-Петербурга и не центра, для локаций в области - 
# медианой по локации

med = data[(data.locality_name == 'Санкт-Петербург')&(data.km_to_center <= 5)].ceiling_height.median()
med_ = data[(data.locality_name == 'Санкт-Петербург')&(data.km_to_center > 5)].ceiling_height.median()    

def potolki_piter(row):
    if row['km_to_center'] <= 5:
        return med
    else:
        return med_                    
                 
data.loc[(data.locality_name == 'Санкт-Петербург')&(data.ceiling_height.isna()), 'ceiling_height'] = \
      data[(data.locality_name == 'Санкт-Петербург')&(data.ceiling_height.isna())].apply(potolki_piter, axis=1)
```

```python
for x in data[(data.locality_name != 'Санкт-Петербург')&(data.ceiling_height.isna())].locality_name.unique():
    
    data.loc[(data.locality_name != 'Санкт-Петербург')&(data.ceiling_height.isna()), 'ceiling_height'] = \
    data[(data.locality_name != 'Санкт-Петербург')&(data.locality_name == x)].ceiling_height.median()
```

```python
# заполним пропуски в данных по жилой площади и площади кухни в зависимости от количества комнат

for x in np.sort(data.rooms.unique())[1:]:
    a = data[data.rooms == x].total_area.median()
    b = data[data.rooms == x].kitchen_area.median()
    c = data[data.rooms == x].living_area.median()
    
    data.loc[(data.rooms == x)&(data.kitchen_area.isna()), 'kitchen_area'] = \
                            (b/a)*data.loc[(data.rooms == x)&(data.kitchen_area.isna()), 'total_area'] 
    data.loc[(data.rooms == x)&(data.living_area.isna()), 'living_area'] = \
                            (c/a)*data.loc[(data.rooms == x)&(data.living_area.isna()), 'total_area']  
```

```python
# удалим также из массива 1 объект в черте города с явно неверной ценой
data[data.last_price < 0.2]
```

```python
data.drop(data[data.last_price < 0.2].index, inplace=True)
```

## 3. Расчёты и добавление результатов

```python
# Добавим колонки с днем недели, месяцем и годом размещения объявления

data['weekday_exposition'] = data['first_day_exposition'].dt.weekday
data['month_exposition'] = data['first_day_exposition'].dt.month
data['year_exposition'] = data['first_day_exposition'].dt.year

def floor_cat(row):
    floors_tot = row['floors_total']
    floor_ = row['floor']
    if floor_ == 1:
        return 'первый'
    elif floor_ == floors_tot:
        return 'последний'
    else:
        return 'другой' 
    
#категоризация по этажам 
data['floor_category'] = data.apply(floor_cat, axis = 1)

# добавим колонки с соотношением жилой и общей площади, соотношением площади кухни и общей
data['living_area_part'] = data['living_area']/data['total_area']
data['kitchen_area_part'] = data['kitchen_area']/data['total_area']

data['1sqm_price'] = data['1sqm_price'].round(3)
data['living_area_part'] = data['living_area_part'].round(2)
data['kitchen_area_part'] = data['kitchen_area_part'].round(2)
```

```python
def gr_floors(h, g, z):
    fig = plt.figure(figsize=(15,3))
    ax1 = fig.add_subplot(1,3,1)
    ax2 = fig.add_subplot(1,3,2)
    ax3 = fig.add_subplot(1,3,3)

    x1 = h.index
    y1 = h
    ax1.plot(x1, y1, marker='o', color = 'red')
    ax1.spines['top'].set_visible(False)
    ax1.spines['right'].set_visible(False)
    ax1.set_ylabel('млн руб')
    ax1.set_ylim([3.5, 5.5])

    x2 = g.index
    y2 = g
    ax2.plot(x2, y2, marker='o', color = 'green')
    ax2.spines['top'].set_visible(False)
    ax2.spines['right'].set_visible(False)
    ax2.set_ylabel('млн руб')
    ax2.set_ylim([4.5, 6])

    x3 = z.index
    y3 = z
    ax3.plot(x3, y3, marker='o', color = 'black')
    ax3.spines['top'].set_visible(False)
    ax3.spines['right'].set_visible(False)
    ax3.set_ylabel('млн руб')
    ax3.set_ylim([2.5, 4])

    ax1.set_title('Медиана цены от этажа, все объекты')
    ax1.title.set_size(12)
    ax2.set_title('Медиана цены от этажа, С-Пб')
    ax2.title.set_size(12)
    ax3.set_title('Медиана цены от этажа, ЛО')
    ax3.title.set_size(12)
```

```python
# создадим стообец с с категорией локации

def cat(row):
    if row['locality_name'] == 'Санкт-Петербург':
        if row['km_to_center'] <= 3:
            return 'C-Пб, центр'
        elif 3 < row['km_to_center'] < 6:
            return 'C-Пб, 3-6 от ц'
        elif 6 <= row['km_to_center'] < 12:
            return 'C-Пб, 6-12 от ц'
        elif row['km_to_center'] >= 12:
            return 'C-Пб, > 12 км от ц'
    else:
        if row['km_to_center'] < 20:
            return 'ЛО, до 20 км'
        elif 20 <= row['km_to_center'] <= 40:
            return 'ЛО, 20-40 км'
        elif row['km_to_center'] > 40:
            return 'ЛО, > 40 км'
        
data['location_cat'] = data.apply(cat, axis=1)

data.groupby('location_cat').last_price.count()
data.groupby('location_cat').last_price.median()

stat_location = pd.DataFrame({
     'кол-во объектов': data.groupby('location_cat').last_price.count(),
    'цена всех объектов': data.groupby('location_cat').last_price.sum().round(2),
     'медиана цены': data.groupby('location_cat').last_price.median().round(2),
     'min цена': data.groupby('location_cat').last_price.min().round(2),
     'max цена': data.groupby('location_cat').last_price.max().round(2),
     'медиана площади': data.groupby('location_cat').total_area.median().round(2),
     'медиана цены за кв м': data.groupby('location_cat')['1sqm_price'].median()
    })


stat_location
```

Массив содержит около 24 тысячи объявлений (66% приходятся на Санкт-Петербург), которые неравномерно распределны по году размещения - менее процента приходится на 2014 год (по году данные только за 2 последних месяца), на 2015-2018 гг. приходится соответственно 5%, 12%, 35%, 36% и на пять месяцев 2019 года - 12%.<br>

95% объектов - помещения стоимостью от 1,2 до 18,4 млн руб, медиана цены составляет 4,65 млн руб, а стоимость 50% квартир находится в диапозоне 3,4-6,8 млн руб.<br>
Диапозон 95% предложений по площади - это 26-126 кв м, чаще предлагались объекты 40-70 кв м, медиана площади составляет 52 кв м.<br>
97% квартир - это 1-3-х комнатные. Высота потолков у 95% объектов находится в диапозоне 2,5-3,1м, медиана - 2,65м.<br>
95% квартир удалены от центра на 2-110 км, 50% - 10-22 км, медиана - 15 км.<br>

После 2014 года наблюдается снижение цены, а в 2019 прослеживается позитивный тренд. Особенно резко выражено падение цены для центра города - цена упала в 3 раза.<br>
Зависимость цены от даты размещения (день недели, месяц) не прослеживается.<br> 
Стоимость объектов не имеет выраженной линейной зависимости ни от одного из признаков, кроме очевидных (площадь и производные). При том, что в целом удаленные от центра объекты чаще более доступны, в ряде локаций вне черты города стоиомость недвижимости достаточно высока - медиана цены в г.Пушкин близка к значению по Санкт-Петербургу (5,5 и 5,2 млн руб).<br>
Последний и первый этажи более доступны, однако в Санкт-Петербурге в целом последний этаж ценится больше, чем другие.
А вот центр города тут повторяет общую тенденцию.<br>
Центр города ожидаемо про высокие цены, однако он не про рекорды - объекты самые дорогие, большие, многокомнатные распложены не здесь. Высокая стоимость квартир в центре прежде всего обусловлена большим метражом - медиана цены 9,8 млн руб, площади - 85 кв метров (значения по всем объектам - 4,65 млн руб и 52 кв м). При цене большей более чем в 2 раза квартиры и на 64% просторнее. Для центра также характерна большая зависимость цены от стоимости квадратного метра и несколько меньшая от площади - если выбор делается между объектом в центре и более удаленным при равной стоиомсти, покупатель поступается метражом ради локации.
